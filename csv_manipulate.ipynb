```python
import csv
import os

def contar_linhas_dados(nome_arquivo: str) -> int:
    """
    Conta quantas linhas de dados (excluindo o cabeçalho) existem
    em um arquivo .txt ou .csv com delimitador ';'.

    Argumentos:
        nome_arquivo (str): O caminho para o arquivo.

    Retorna:
        int: O número de linhas de dados, ou -1 em caso de erro.
    """
    try:
        with open(nome_arquivo, mode='r', encoding='ISO-8859-1', newline='') as f:
            # 1. Cria um "leitor" CSV, especificando o delimitador
            leitor = csv.reader(f, delimiter=';')

            # 2. Tenta pular a primeira linha (cabeçalho)
            try:
                next(leitor)
            except StopIteration:
                # O arquivo está completamente vazio (nem cabeçalho tem)
                return 0

            # 3. Conta as linhas restantes de forma eficiente
            contador_linhas = sum(1 for linha in leitor)

            return contador_linhas

    except FileNotFoundError:
        print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado.")
        return -1
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return -1

def ler_dados_completos(nome_arquivo: str):
    """
    Lê e exibe o cabeçalho e os 5 primeiros dados em formato de tabela.
    """
    print(f"\n--- Lendo dados de '{nome_arquivo}' ---")
    try:
        with open(nome_arquivo, mode='r', encoding='ISO-8859-1', newline='') as f:
            leitor = csv.reader(f, delimiter=';')

            # Pega o cabeçalho
            try:
                cabecalho = next(leitor)
            except StopIteration:
                print("Arquivo está vazio.")
                return

            # Itera sobre os dados para pegar a amostra
            amostra_dados = []
            for i, linha in enumerate(leitor):
                if i == 300 :  # para depois de 5 linhas
                    break
                amostra_dados.append(linha)

            # --- INÍCIO DA LÓGICA DA TABELA ---

            if not amostra_dados:
                print(f"**Cabeçalho:** {cabecalho}")
                print("Não há linhas de dados no arquivo.")
                return

            # 1. Calcular a largura máxima de cada coluna
            num_colunas = len(cabecalho)
            # Começa com a largura do próprio cabeçalho
            larguras = [len(str(col)) for col in cabecalho]

            # Atualiza as larguras com base nos dados da amostra
            for linha in amostra_dados:
                for i in range(num_colunas):
                    # Checa se a linha de dados tem essa coluna (evita erros)
                    if i < len(linha):
                        larguras[i] = max(larguras[i], len(str(linha[i])))

            # 2. Criar a string de formato para cada linha
            # O ' ' extra é para dar um espaçamento
            formato_linha = " | ".join([f"{{:<{w}}}" for w in larguras])

            # 3. Imprimir o Cabeçalho formatado
            print(f"**Cabeçalho e {len(amostra_dados)} primeiras linhas:**")
            print(formato_linha.format(*cabecalho))

            # 4. Imprimir o Separador
            print("-+-".join(["-" * w for w in larguras]))

            # 5. Imprimir as linhas de dados formatadas
            for linha in amostra_dados:
                # Garante que a linha tenha o n° de colunas para o .format()
                if len(linha) < num_colunas:
                    # Preenche colunas faltantes com ''
                    linha.extend([''] * (num_colunas - len(linha)))
                print(formato_linha.format(*linha))

            # --- FIM DA LÓGICA DA TABELA ---

    except FileNotFoundError:
        print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado.")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")


# --- Bloco de Execução Principal ---
if __name__ == "__main__":

    arquivo_para_ler = "enade_compilado.csv" # Mantido seu arquivo

    # 1. Executa a função que você pediu
    print("Iniciando contagem de linhas...")
    total_dados = contar_linhas_dados(arquivo_para_ler)

    if total_dados != -1:
        print(f"Resultado: O arquivo contém {total_dados} linhas de dados (fora o cabeçalho).")

    print("-" * 10)

    # 2. Executa a função bônus para mostrar a leitura
    ler_dados_completos(arquivo_para_ler)
```
